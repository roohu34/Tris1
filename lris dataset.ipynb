```python
x=a[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].fillna(1)
x
```

```python
y=a[["Species"]]
y
```

```python
y=a.replace[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"],[0,1,0]]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=2,shuffle=True)
```

```python
from sklearn.svm import SVC
```

```python
support=svm.SVC(kernel="linear",random_state=2)
```

```python
support.fit(xtrain,ytrain)
```

```python
support.predict(xtest)
```

```python
support.score(xtest,ytest)
```
